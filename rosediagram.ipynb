Rose diagrams 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mplstereonet
```

```python
df=pd.read_excel('ripple-crest.xlsx')
strikes=df['ripples'].tolist()
dips=df['dips'].tolist()
```

```python
bin_edges = np.arange(-5, 366, 10)
bin_edges
number,bin_edges=np.histogram(strikes,bin_edges)
number
```

```python
bin_edges_d = np.arange(-5, 366, 10)
number_d,bin_edges_d=np.histogram(dips,bin_edges_d)
number_d
```

```python
number[0] += number[-1]
half = np.sum(np.split(number[:-1], 2), 0)
two_halves = np.concatenate([half, half])
```

```python
fig = plt.figure(figsize=(14,7))

ax = fig.add_subplot(122, projection='polar')

ax.bar(np.deg2rad(np.arange(0, 360, 10)), two_halves, 
       width=np.deg2rad(10), bottom=0.0, color='.8', edgecolor='k')
ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)
ax.set_thetagrids(np.arange(0, 360, 10), labels=np.arange(0, 360, 10))
ax.set_rgrids(np.arange(1, two_halves.max() + 1, 2), angle=0, weight= 'black')
ax.set_title('Rose Diagram of Ripple Crest Strikes', y=1.10, fontsize=15)

fig.tight_layout()
```

```python
number_d[0] += number_d[-1]
half = np.sum(np.split(number_d[:-1], 2), 0)
two_halves_d = np.concatenate([half, half])
number_d
```

```python
fig = plt.figure(figsize=(14,7))

ax = fig.add_subplot(122, projection='polar')

ax.bar(np.deg2rad(np.arange(0, 360, 10)), two_halves_d, 
       width=np.deg2rad(10), bottom=0.0, color='.8', edgecolor='k')
ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)
ax.set_thetagrids(np.arange(0, 360, 10), labels=np.arange(0, 360, 10))
ax.set_rgrids(np.arange(1, two_halves_d.max() + 1, 2), angle=0, weight= 'black')
ax.set_title('Rose Diagram of the "Fault System"', y=1.10, fontsize=15)

fig.tight_layout()
```
